```python
import sqlite3
```

```python
conn=sqlite3.connect("bookstore.db")
```

```python
cursor=conn.cursor()
```

books Table


```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS books (
    book_id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT NOT NULL,
    author TEXT NOT NULL,
    price REAL NOT NULL,
    stock_quantity INTEGER DEFAULT 0
)
''')
```

customers tables

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS Customers  (
    customer_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    EMAIL TEXT  UNIQUE NOT NULL,
    city TEXT,
    join_date TEXT
)
''')
```

 Orders Table:



```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS Orders  (
    order_id INTEGER PRIMARY KEY AUTOINCREMENT,
    customer_id INT,
    book_id INT,
    quantity INTEGER  NOT NULL,
    order_date  TEXT NOT NULL,
    TOTAL_AMOUNT REAL,
    FOREIGN KEY(customer_id ) REFERENCES CUSTOMERs(customer_id ),
    FOREIGN KEY (book_id)  REFERENCES BOOKs(book_id)
)
''')
```

```python
#Find the customer who has placed the most orders (use GROUP BY and ORDER BY)
cursor.execute('''
select customers.name,count(orders.order_id) as most_orders from customers 
               left join orders
               on  customers.customer_id = orders.customer_id
               group by customers.customer_id
               order by  most_orders desc
               limit 2;
              
               
''')
print(cursor.fetchall())
```

```python
query='''select orders.order_id,customers.name, customers.City,books.title,
Orders.Quantity,Orders.TOTAL_AMOUNT from customers
 left join orders on
   customers.customer_id = orders.customer_id 
 left join books on 
    books.book_id = orders.book_id'''
report=pd.read_sql(query,conn)
print(report)
```

Calculate:
Average order value,
Total orders by city,
Most popular book (most ordered)

```python
query1 = """
SELECT AVG(total_amount) AS average_order_value
FROM Orders
"""
report1 = pd.read_sql(query1, conn)
print(report1)
```

```python
query2 = """
SELECT 
    customers.city,
    COUNT(orders.order_id) AS total_orders_by_city
FROM customers
LEFT JOIN orders 
    ON customers.customer_id = orders.customer_id
GROUP BY customers.city
"""
report2 = pd.read_sql(query2, conn)
print(report2)
```

```python
#Most popular book (most ordered)
query3='''SELECT b.title, COUNT(o.order_id) AS total_orders
FROM books b
LEFT JOIN orders o
ON b.book_id = o.book_id
GROUP BY b.book_id
ORDER BY total_orders DESC
LIMIT 1;'''
report3 = pd.read_sql(query3, conn)
print(report3)
```

Part B: Pandas to SQL



```python
#Create a new DataFrame with discount information:
import pandas as pd
discounts_data = {
    'book_id': [1, 2, 3, 4, 5],
    'discount_percent': [10, 15, 5, 20, 12]
}
discounts=pd.DataFrame(discounts_data)
discounts
```

```python
discounts.to_sql('discounts', conn, if_exists='replace', index=False)
```

```python
cursor.execute('''select * from discounts''')
print(cursor.fetchall())
```

Write a SQL query joining books and discounts to show book titles with their discounted prices


```python
q='''select books.title ,books.price as original_price,discounts.discount_percent,  books.price - (books.price * discount_percent / 100) AS discounted_price
 from books 
left join discounts on 
books.book_id = discounts.book_id'''
disc=pd.read_sql(q,conn)
print(disc)
```
